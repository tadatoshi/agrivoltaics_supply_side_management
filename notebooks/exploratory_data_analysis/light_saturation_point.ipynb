```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import arviz as az

import pymc3 as pm
```

```python
from scipy import stats
```

```python
with pm.Model() as simple_model:
    light_saturation_point = pm.Normal("light_saturation_point", mu=350, sigma=1)
    
    prior = pm.sample_prior_predictive()
```

```python
prior
```

```python
prior['light_saturation_point'].shape
```

```python
plt.plot(prior['light_saturation_point'])
```

```python
plt.hist(prior['light_saturation_point'], density=True, alpha=0.3)
```

```python
light_saturation_point_kde = stats.gaussian_kde(prior['light_saturation_point'])
```

```python
light_saturation_point_kde
```

```python
ordered_light_saturation_points = np.sort(prior['light_saturation_point'])
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.hist(prior['light_saturation_point'], density=True, alpha=0.3)
ax.plot(ordered_light_saturation_points, light_saturation_point_kde(ordered_light_saturation_points))
```
